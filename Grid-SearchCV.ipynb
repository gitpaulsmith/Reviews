```python
import pylab as pl
import numpy as np
import string
%pylab inline
import pandas as pd
pd.options.display.mpl_style = 'default'
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.pipeline import Pipeline
from sklearn.cross_validation import ShuffleSplit, train_test_split 
from sklearn.metrics import precision_recall_curve, auc, confusion_matrix
```

```python
fn = 'foods.csv'
f = open(fn,'r')
df = pd.io.parsers.read_table(f, '\t!\t',engine='python')
f.close()
df=df.drop_duplicates('text')
df=df.replace({'score': {'5.0':5,'4.0':4,'3.0':3,'2.0':2,'1.0':1}})
df_24stars=pd.DataFrame(df[(df['score']==2) | (df['score']==4)],columns=['score','text'])
df_15stars=pd.DataFrame(df[(df['score']==1) | (df['score']==5)],columns=['score','text'])
```

```python
def create_ngram_model():
    tfidf_ngrams=TfidfVectorizer(ngram_range=(1,3),analyzer="word",binary=False)
    clf=MultinomialNB()
    pipeline=Pipeline([('vect',tfidf_ngrams),('clf',clf)])
    return pipeline
```

```python
def train_model(clf,X,Y):

     # setting random_state to get deterministic behavior
    #cv = ShuffleSplit(n=len(X), n_iter=10, test_size=0.3,indices=True, random_state=0)
    train_size=0.6
    X_train=X[:train_size*X.size]
    y_train=X[:train_size*Y.size]
    
    X_test=X[train_size*X.size:]
    y_test=X[train_size*Y.size:]
    
    scores = []
    pr_scores = []

    
    #clf = clf_factory()
    clf.fit(X_train, y_train)

    train_score = clf.score(X_train, y_train)
    test_score = clf.score(X_test, y_test)

    scores.append(test_score)
    proba = clf.predict_proba(X_test)
    precision, recall, pr_thresholds = precision_recall_curve(y_test, proba[:,1])

    pr_scores=auc(recall, precision)
    
    summary = (np.mean(scores), np.std(scores),
               np.mean(pr_scores), np.std(pr_scores))
    print "%.3f\t%.3f\t%.3f\t%.3f"%summary
```

```python
from sklearn.grid_search import GridSearchCV
from sklearn.metrics import f1_score

def grid_search_model(clf_factory, X, Y):
    #cv = ShuffleSplit(
        #n=len(X), n_iter=10, test_size=0.3, indices=True, random_state=0)
    X_train, X_test, y_train, y_test= train_test_split(X, Y, test_size=0.6, random_state=0)
    
    param_grid = dict(vect__ngram_range=[(1, 1), (1, 2), (1, 3)],
                      vect__min_df=[1, 2],
                      vect__stop_words=[None, "english"],
                      vect__smooth_idf=[False, True],
                      vect__use_idf=[False, True],
                      vect__sublinear_tf=[False, True],
                      vect__binary=[False, True],
                      clf__alpha=[0, 0.01, 0.05, 0.1, 0.5, 1],
                      )

    grid_search = GridSearchCV(clf_factory(),
                               param_grid=param_grid,
                               score_func=f1_score,
                               cv=2,
                               verbose=10)
    grid_search.fit(X_train,y_train) 
 
    return grid_search.best_estimator_
```

```python
def makePR_Plot(x,y,title):
    pl.figure
    pl.plot(x, y)
    pl.xlabel('Recall')
    pl.ylabel('Precision')
    pl.ylim([0.0, 1.05])
    pl.xlim([0.0, 1.0])
    pl.title(title)
    pl.fill_between(x,0,y)
    pl.savefig(title)
```
