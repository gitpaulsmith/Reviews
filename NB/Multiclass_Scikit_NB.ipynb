#### Loading the data from the csv file

```python
fn = 'foods.csv'
f = open(fn,'r')
df = pd.io.parsers.read_table(f, '\t!\t',engine='python')
f.close()
df=df.drop_duplicates('text')
df=df.replace({'score': {'5.0':5,'4.0':4,'3.0':3,'2.0':2,'1.0':1}})
df_24stars=pd.DataFrame(df[(df['score']==2) | (df['score']==4)],columns=['score','text'])
df_15stars=pd.DataFrame(df[(df['score']==1) | (df['score']==5)],columns=['score','text'])
df_3stars=pd.DataFrame(df[(df['score']==3)],columns=['score','text'])
df_allstars=pd.DataFrame(df,columns=['score','text'])
```

#### ngram TF-IDF Vectorizer function, nrange is the number of words in the bundle

```python
def create_ngram_model(nrange,stopwords):
    tfidf_ngrams=TfidfVectorizer(ngram_range=(1,nrange),analyzer="word",binary=False,stop_words=stopwords)
    clf=MultinomialNB()
    pipeline=Pipeline([('vect',tfidf_ngrams),('clf',clf)])
    return pipeline
```

#### Training and testing functions based on Naive Bayes model

```python
def just_train_NB(clf_factory,*args):
    X_train=args[0]
    Y_train=args[1]
    n=args[2]
    scores=[]
    
    clf=clf_factory(n,args[3])
    clf.fit(X_train,Y_train)
    
    train_score=clf.score(X_train,Y_train)
    print 'Train Score is ',train_score
    
    return clf

def just_predict(clf,X_test,Y_test):
    pr_scores=[]
    y_pred=clf.predict(X_test)
    proba=clf.predict_proba(X_test)
    print proba.shape
    print Y_test.shape
    
    #precision,recall,pr_thresholds=precision_recall_curve(Y_test,proba)
    #pr_scores.append(auc(recall,precision))
    
    cm=confusion_matrix(Y_test,y_pred)
    cm=cm / cm.astype(np.float).sum(axis=1)
    print cm

    summary = (np.mean(pr_scores), np.std(pr_scores))
    print "%.3f\t%.3f"%summary
    return cm
```

#### Function to draw the Precision-Recall plot

```python
def makePR_Plot(x,y,title,n):
    pl.figure(n)
    pl.plot(x, y)
    pl.xlabel('Recall')
    pl.ylabel('Precision')
    pl.ylim([0.0, 1.05])
    pl.xlim([0.0, 1.0])
    pl.title(title)
    pl.fill_between(x,0,y)
    pl.savefig(title)
    
def makeConfusion_Matrix_Plot(cm,title,n):
    pl.figure(n)
    pl.matshow(cm)
    pl.title(title)
    pl.colorbar()
    pl.ylabel('True label')
    pl.xlabel('Predicted label')
    pl.show()
    pl.savefig(title)
```

#### Copying a sample of the one and five start data to X and Y numpy arrays

```python
def decodeArray(text):
    try:
        return text.decode("utf-8","replace")
    except:
        return 0
```

#### Checking the number of 'negative' and 'positive' reviews

```python
sampleSize=10000

pieces=[df_allstars[df_allstars['score']==5][:sampleSize],df_allstars[df_allstars['score']==4][:sampleSize],
        df_allstars[df_allstars['score']==3][:sampleSize],df_allstars[df_allstars['score']==2][:sampleSize],
        df_allstars[df_allstars['score']==1][:sampleSize]]
final_df=pd.concat(pieces)
final_df=final_df.reindex(np.random.permutation(final_df.index))
X=array(final_df['text'].apply(decodeArray))
Y=array(final_df['score'])
```

```python
classes = np.unique(Y)
print classes
for c in classes:
        print("#%s: %i" % (c, sum(Y==c)))
```

```python
test_size=0.3
test_start=round(test_size*sampleSize)
X[test_start:].shape
```

```python
best_cm=just_predict(best_pipeline,X[:test_start],Y[:test_start])
```
